```
df=pd.read_csv('wages.csv')
df.head()
```

```
df.sex.unique()
```

```
df['sex_code'] = df.sex.map({'male':0,'female':1})
df.head()
```

```
df.sex.replace({'male':0,'female':1},inplace=True) # inplace=True가 없으면 저장안됨
# = df.sex.replace(['male','female'],[0,1])
df
```

```
del df['sex_code']
```

```
df
```

```
df_info = df[['earn','height','age']]
df_info
```

```
f = lambda x:x.max() - x.min()
df_info.apply(f) # apply는 dataframe을 대상으로
```

```
df_info.age.map(lambda x:x*2) # 위의 apply는 series에서도 가능하며
# map은 series에서만 가능함, 즉 여기서 map대신 apply써도 동일
```

```
def f(x):
  return Series([x.min(), x.max()], index = ['min', 'max'])
df_info.apply(f)
```

```
f = lambda x: -x
df_info.applymap(f).head(3) # applymap은 모든 element를 대상으로! dataframe에서만 사용가능
```

```
df_info.age.apply(f).head(3) # series이므로 apply, map 둘다 사용가능
```
