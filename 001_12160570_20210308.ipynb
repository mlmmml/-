```
x1 = 100
y1 = 100
print(id(x1), id(y1))
x2 = 300
y2 = 300
print(id(x2), id(y2))

# 출력 전 x1과 y1, 그리고 x2와 y2는 입력된 값이 같더라도 서로 다른 변수이기 때문에 주소가 다르게 나올 것이라 예상했다.
# 하지만 출력 결과 x1과 y1의 주소는 같다고 나온 반면, x2와 y2의 주소는 다르게 나왔다.
# x1과 y1의 출력 결과로 보아 변수에 입력된 값을 기준으로 주소가 할당되었다고 이해할 수 있다.
# 하지만 x2와 y2의 출력 결과를 보면 입력된 값이 같음에도 불구하고 주소가 다르게 나왔으므로 변수를 기준으로 주소가 할당되었다고 이해할 수 있다.
# 이해를 위해 반복해서 코드를 돌려보았는데, x1과 y1의 주소는 항상 똑같이 나오는 반면, x2와 y2의 주소는 계속해서 바뀌는 것을 알 수 있었다.
```
