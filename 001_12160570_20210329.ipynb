```python
f = open('yesterday.txt', 'r')
yesterday_lyric = ""
while 1:
  line = f.readline()
  if not line:
    break
  yesterday_lyric = yesterday_lyric + line
f.close()
yesterday_list = yesterday_lyric.split()
j = 0
for i in range(len(yesterday_list)):
  if yesterday_list[i] == "Yesterday":
    j += 1
print(j)
```

```python
j = 0
for i in range(len(yesterday_list)):
  if yesterday_list[i] == "yesterday":
    j += 1
print(j)
```

```python
j = 0
for i in range(len(yesterday_list)):
  if yesterday_list[i] == "Yesterday" or yesterday_list[i] == "yesterday":
    j += 1
print(j)
```

```python
def txtchange(x):
  y = x.split('.')
  print('문장의 수 :', len(y))
  z = x.replace(' ','')
  print('글자의 수 :', len(z))
text = input('입력된 문자열 : ')
txtchange(text)
```
