```
import pandas as pd
sample=pd.read_csv('CMAPSS_sample.txt')
sample.head()
```

```
sample=pd.read_csv('CMAPSS_sample.txt',sep=' ',header=None) #header=None : column값이 없다는 의미
sample.head()
```

```
sample.drop([26, 27],axis=1, inplace=True) #axis=1까지만 하면 drop된 상태를 보여주기만할뿐 다시 sample 출력하면 그대로임. 따라서 inplace=True 필요. (저장느낌)
sample.head()
```

```
a=['UnitNumber','Cycle']
b=['Op_Setting_'+str(i) for i in range(1,4)] # str(i) : 문자열i
c=['Sensor_'+str(i)for i in range(1,22)]
d=a+b+c
sample.columns=d # sample.columns=a+b+c라 해도됨
sample.head()
```

```
sample['UnitNumber']==77 #바로 아래보면 알수있듯이 9920부터 10081까지 77이므로 이 범위내에선 True값이 나올것
```

```
sample[sample['UnitNumber']==77]
```

```
sample_77=sample[sample['UnitNumber']==77]
sample_77
```

```
df=sample_77.describe()
df
```

```
s66=sample[sample.UnitNumber==65] #그냥 확인차 넣어봄
s66
```

```
series=sample.groupby('Cycle')['Sensor_11'].mean() #Cycle을 기준으로 Sensor_11의 평균을 구함
#아래에서 보면 알수있듯이 UnitNumber는 100까지있고, 각UnitNumber엔 Cycle의 수가 다 다르게 있음
series
```

```
max_cycle=sample.groupby('UnitNumber')['Cycle'].max()
max_cycle
```

```
max_cycle=sample.groupby('UnitNumber')['Cycle'].max().reset_index()
max_cycle
```

```
max_cycle[max_cycle.UnitNumber==77] #위에 sample_77참고하고 162맞는지 확인하라고.
```

```
sample
```

```
df=pd.merge(max_cycle,sample,on='UnitNumber',how='inner')
df #Cycle_x:max_cycle의 Cycle에 의한 각UnitNumber별 Cycle의 최댓값, Cycle_y:sample의 Cycle값
```

```
sample['Remaning_Life']=df['Cycle_x']-sample['Cycle'] #새로운 Series 생성
sample
```
