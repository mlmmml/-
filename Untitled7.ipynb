```
import pandas as pd
```

```
df = pd.read_csv('wages.csv')
df
```

```
for i in range(len(df)):
  df['psc'].iloc[i]='num'+str(i)
df
```

```
df.race.unique()
```

```
df.groupby('race').filter(lambda x: len(x)<127)
```

```
df.groupby('race').filter(lambda x: x['height'].mean()<66)
```

```
k1 = df[df['height']>70]
c1 = k1[['sex','age','psc']]
c1
```

```
k2 = df[df['age']<35]
c2 = k2[['race','psc']]
```

```
pd.merge(c1,c2,on='psc') 
```
