```
df=pd.read_excel('통합 문서2.xlsx',header=1)
a1=list(df['엘리베이터 기종'].unique()) # 기종이 뭐가있나 파악
for i in range(len(a1)):
  df1=df[df['엘리베이터 기종']==a1[i]] # 분석하고자하는 기종선택 ★
  d1=df1.groupby('고장내용')['처리결과'].count()
  nub=1
  print('◆ 기종['+a1[i]+']')
  print(nub,'. 고장내용 비율 상위 5개\n')
  z=dict(round(d1/d1.sum()*100,2))
  z1=dict(sorted(z.items(), key=lambda x:x[1], reverse=True)) # 선택한 기종의 고장내용 비율, 내림차순 정렬
  xx1=[]
  yy1=[]
  bb=0
  for a,b in z1.items():
    bb+=1
    print(nub,'- (',bb,') %s : %s'%(a,b),'%',end='')
    if b>10: print(' ☜')
    else : print('')
    xx1.append(a)
    yy1.append(b)
    if len(yy1)==5:  # 상위 5개 표시하겠다!
      break

  # plt.bar(xx1,yy1)
  # for nj in range(len(xx1)):
  #   plt.text(nj-0.2,yy1[nj]+0.4,yy1[nj],size=10)
  # plt.show() # ★★


  y22=[]
  x22=[]
  nub+=1
  for bm in range(len(yy1)):
    if yy1[bm]>=10:      # 고장내용 비율 10% 이상만
      x22.append(xx1[bm])
      y22.append(yy1[bm])
  print('\n',nub,'. 10% 이상인 고장내용')
  for jk in range(len(x22)):
    
    print(end='')
  
  for j in range(len(y22)): 
    k=len(df1['엘리베이터 호기'].unique())
    q=list(df1['엘리베이터 호기'].unique())
    p=[]
    w=0
    print('\n',nub,'- (',j+1,')',xx1[j],':',y22[j],'%\n')
    print('호기별 ['+xx1[j]+']가 차지하는 비율')
    for r in range(k):
      dfpp=df1[df1['엘리베이터 호기']==q[r]]
      print(q[r],':',round(dfpp[dfpp['고장내용']==xx1[j]]['엘리베이터 기종'].count()/df1[df1['엘리베이터 호기']==q[r]]['고장내용'].count()*100,2),'%  (',dfpp[dfpp['고장내용']==xx1[j]]['엘리베이터 기종'].count(),'/',df1[df1['엘리베이터 호기']==q[r]]['고장내용'].count(),')')
      ###############↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑
      p.append(dfpp[dfpp['고장내용']==xx1[j]]['엘리베이터 기종'].count()/df1[df1['엘리베이터 호기']==q[r]]['고장내용'].count()*100)
    for t in range(len(p)):
      w+=p[t]
    print('-> 호기별 ['+xx1[j]+']가 차지하는 비율 평균:',round(w/len(p),2),'%\n') # ★★★, ★★★★
    g1=[]
    hk=[]
    print('['+xx1[j]+']에 대한 조치부품별 정상작동시간')
    yu=[]
    bg=[]
    fr=0
    for w1 in range(k):
      dfp=df1[df1['엘리베이터 호기']==q[w1]]
      #print(q[w1]) #####################

      dfp2=dfp[dfp['고장내용']==xx1[j]]
      g=[]
      for w2 in range(len(dfp2)-1):
        g.append(dfp2.iloc[w2,5])
        g1.append(dfp2.iloc[w2,5])
      for e5 in range(len(dfp2)): 
        bg.append(dfp2.iloc[e5,5]) 
        
      h=[] 
      for w3 in range(len(dfp2)-1):
        h.append(dfp2.iloc[w3,6])

      m=list(dfp[dfp['고장내용']==xx1[j]]['고장일시'])
      m1=[]

      for w4 in range(len(m)):
        m1.append(dt.strptime(m[w4],'%Y-%m-%d %H:%M:%S'))
      for w5 in range(len(m1)-1): ##############
        print('조치부품 [',g[w5],'], 처리결과 [',h[w5],'] :',str(m1[w5+1]-m1[w5])) ####################
        hk.append(g[w5])
    
      hk1=list(set(hk)) # 조치부품 unique
      for ti in range(len(g)):
        yu.append(g[ti])
      
      
    tu=0
    sf={}
    fg=[]
    
    for w14 in range(k):
      dfp=df1[df1['엘리베이터 호기']==q[w14]]

      o=list(dfp[dfp['고장내용']==xx1[j]]['고장일시'])
    
      e1=[]
      for w7 in range(len(o)):
        e1.append(dt.strptime(o[w7],'%Y-%m-%d %H:%M:%S'))
      
      for uh in range(len(hk1)):
        if fr ==len(hk1):
          break
        else:
          sf[hk1[uh]]=[]
          fr+=1
      
      for y5 in range(len(e1)-1):
        fg.append((e1[y5+1]-e1[y5]))


    for y6 in range(len(hk)):
      if tu==len(hk):
        break
      else:
        tu+=1
        sf[yu[y6]].append(fg[y6])

    #print(sf)


      #print() # ★★★★★ #################

    bg1=list(set(bg))
    hh=[]
    kk={}
    for h2 in range(len(bg1)):
      for h1 in range(len(bg)):
        if bg[h1]==bg1[h2]:
          hh.append(bg[h1])
          k4=bg1[h2]
      k5=round(len(hh)/len(bg)*100,2) # g2[h2]의 비율
      kk[k4]=k5 
      hh=[]
    kk3=dict(sorted(kk.items(), key=lambda x:x[1], reverse=True))  # 가장많이 사용된 조치부품 내림차순 정렬
    kk4={}
    #print(kk3) #확인차
    nb=1
    for a6,b6 in kk3.items():
      if b6>10:
        kk4[a6]=b6
    print('\n',nub,'- (',j+1,') -',nb,'. 기종['+a1[i]+']에서 발생한 고장['+xx1[j]+']에 사용된 조치부품 (10%이상)\n')
    ji=1
    for bf1, bf2 in kk4.items():
      print(ji,'.',bf1,':',bf2,'%')
      ji+=1
    nmx=list(kk4.keys())
    nmy=list(kk4.values())

    # plt.bar(nmx,nmy,color='r')
    # for nj1 in range(len(nmx)):
    #   plt.text(nj1-0.1,nmy[nj1]+0.35,nmy[nj1],size=10)
    # plt.show()

    nb+=1
    print('\n',nub,'- (',j+1,') -',nb,'. 고장['+xx1[j]+']에 사용된 조치부품 중 이후 정상작동시간 평균 상위 5개')
    
    wt=[]
    for ng in range(len(hk1)):
      tt=timedelta(0)
      for hg in range(len(sf[hk1[ng]])):
        tt+=sf[hk1[ng]][hg]/len(sf[hk1[ng]])
      wt.append(tt)
    #print('\n',hk1)
    #print(wt)
    dct={}
    for vx in range(len(hk1)):
      dct[hk1[vx]]=wt[vx]

    dct2=dict(sorted(dct.items(), key=lambda x:x[1], reverse=True))
    print('\n')
    xp1=[]
    yp1=[]
    for qp,qr in dct2.items():
      xp1.append(qp)
      yp1.append(qr)
    #print(xp1)
    #print(yp1)

    fq=0
    hr=1
    for pl in range(len(dct2)):
      if fq==5:
        break
      else:
        print(hr,'.',xp1[pl],': ',end='')
        print(yp1[pl])
        hr+=1
        fq+=1
    
    #plt.bar(xp1,yp1)
    #plt.show()
    print()
    
```
